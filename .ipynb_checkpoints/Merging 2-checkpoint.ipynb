```python
%reset -fs
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas_profiling as pp
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import unidecode
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
df = pd.read_csv('fuzzy_matches_bet_df.csv', index_col=0)
df
```

```python
df2 = pd.read_csv('player_team_stats_merged.csv', index_col=0)
df2
```

```python
matches = df['Matches'].str.split(',', expand=True)
matches
```

```python
matches[0] = matches[0].map(lambda x: x.replace('[','').replace('(', '').replace("'", '')).replace('"', '')
matches[0] = matches[0].map(lambda x: x.strip('"'))
```

```python
matches[matches[0] == 'DAngelo Russell']
```

```python
df['Matches'] = matches[0]
df
```

```python
matches = df.Matches.tolist()
```

```python
matches
```

```python
matches_list = []
for i in matches:
    output = unidecode.unidecode(i)
    matches_list.append(output)
matches_list
```

```python
df['Matches'] = matches_list
```

```python
players = df2.Player.tolist()
players
```

```python
players_list = []
for i in players:
    output = unidecode.unidecode(i)
    players_list.append(output)
players_list
```

```python
df2['Player'] = players_list
```

```python
df2['Player'] = df2['Player'].map(lambda x: x.replace('[','').replace('(', '').replace("'", '')).replace('"', '')
df2['Player'] = df2['Player'].map(lambda x: x.strip('"'))
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
df2['Date'] = pd.to_datetime(df2.Date)
```

```python
merged_df = pd.merge(df, df2,  how='left', left_on=['Date','Matches'], right_on = ['Date','Player'])
merged_df
```
